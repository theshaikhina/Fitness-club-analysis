```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, silhouette_samples
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

### Чтение данных

```python
clients = pd.read_csv('clustering_update.csv')
```

```python
clients.head(5)
```

```python
clients.info()
```

#### Определение признаков

```python
features = ['age', 'gender', 
            'fitness_category_crossfit', 'fitness_category_goboxing',
            'fitness_category_АКАДЕМИЯ_ИЛЬИНА', 'fitness_category_ВОСТОЧНЫЕ_ЕДИНОБОРСТВА',
            'fitness_category_ГРУППОВЫЕ_ПРОГРАММЫ', 'fitness_category_ДЕТСКИЙ_КЛУБ',
            'fitness_category_ТРЕНАЖЕРНЫЙ_ЗАЛ', 'fitness_category_СТУДИЯ_ЙОГИ_ШАКТИ', 
            'fitness_category_ВОДНЫЕ_ПРОГРАММЫ', 'active', 'membership_id']
```

#### Масштабирование признаков

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(clients[features])
```

#### Определение оптимального числа кластеров с помощью метода локтя

```python
inertia = []
silhouette_scores = []
K = range(2, 15)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
    silhouette_avg = silhouette_score(X_scaled, kmeans.labels_)
    silhouette_scores.append(silhouette_avg)
```

####  Построение графика метода локтя

```python
plt.figure(figsize=(10, 6))
plt.plot(K, inertia, 'bo-', label='Inertia')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.legend()
plt.show()
```

#### Построение графика коэффициента силуэта

```python
plt.figure(figsize=(10, 6))
plt.plot(K, silhouette_scores, 'bo-', label='Silhouette Score')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score For Optimal k')
plt.legend()
plt.show()
```

```python
optimal_k = 13
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(X_scaled)
```

```python
clients['cluster'] = clusters
```

```python
# Расчет коэффициента силуэта для каждого объекта
silhouette_vals = silhouette_samples(X_scaled, clients['cluster'])
silhouette_avg = silhouette_score(X_scaled, clients['cluster'])

print(f'Silhouette Score: {silhouette_avg}')
```

```python
# Применение PCA для снижения размерности до 2D
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Создание DataFrame с результатами PCA и кластерами
pca_df = pd.DataFrame(data=X_pca, columns=['PCA1', 'PCA2'])
pca_df['cluster'] = clusters

# Визуализация кластеров на 2D графике
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='cluster', data=pca_df, palette='viridis')
plt.title('PCA of Clients Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
from sklearn.metrics import silhouette_samples
import numpy as np
import matplotlib.cm as cm

# Расчет коэффициента силуэта для каждого объекта
silhouette_vals = silhouette_samples(X_scaled, clients['cluster'])
silhouette_avg = silhouette_score(X_scaled, clients['cluster'])

print(f'Silhouette Score: {silhouette_avg}')

# Визуализация коэффициента силуэта для каждого кластера
y_lower, y_upper = 0, 0
yticks = []

plt.figure(figsize=(10, 6))

for i in range(optimal_k):
    cluster_silhouette_vals = silhouette_vals[clients['cluster'] == i]
    cluster_silhouette_vals.sort()
    y_upper += len(cluster_silhouette_vals)
    plt.barh(range(y_lower, y_upper), cluster_silhouette_vals, height=1.0, edgecolor='none', color=cm.nipy_spectral(float(i) / optimal_k))
    yticks.append((y_lower + y_upper) / 2)
    y_lower += len(cluster_silhouette_vals)

plt.axvline(silhouette_avg, color="red", linestyle="--") 
plt.yticks(yticks, [f'Cluster {i}' for i in range(optimal_k)])
plt.ylabel('Cluster')
plt.xlabel('Silhouette coefficient')
plt.title('Silhouette plot for the clusters')
plt.show()
```
